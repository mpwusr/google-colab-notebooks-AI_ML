```python

import locale

def get_currency(bill):
  locale.setlocale(locale.LC_ALL, 'en_US.UTF-8')
  return locale.currency(bill, grouping=True)


def apply_discount(amount, discount):
  return amount-(amount*discount)


def print_bill(amount):
  if amount > 100.00 and amount < 250.00:
    final_bill = apply_discount(amount, 0.20)
    formatted_bill ='${:,.2f}'.format(final_bill)
    bill_currency = get_currency(final_bill)
    print(f'the bill is a total of {formatted_bill} which is greater than $100.00 before discount')
  elif amount > 250.00:
    final_bill = apply_discount(amount, 0.25)
    formatted_bill ='${:,.2f}'.format(final_bill)
    bill_currency = get_currency(final_bill)
    print(f'the bill is a total of {formatted_bill} which is greater than $250.00 before discount')
  else:
    final_bill = apply_discount(amount, 0.125)
    formatted_bill ='${:,.2f}'.format(final_bill)
    bill_currency = get_currency(final_bill)
    print(f'the bill is a total of {formatted_bill} which is less than $100.00 before discount')

print_bill(99.99)
print_bill(115.07)
print_bill(259.57)
print_bill(14.76)
print_bill(173.45)
print_bill(314.76)
print_bill(173.45)
```

```python
import random
import time

SIZE = 10000
input_data = [0] * SIZE

def create_test_data():
    global input_data
    input_data = [random.randint(0, 10000) for _ in range(SIZE)]

def bubble_sort(arr):
    n = len(arr)
    for i in range(n):
        for j in range(0, n-i-1):
            if arr[j] > arr[j+1]:
                arr[j], arr[j+1] = arr[j+1], arr[j]

if __name__ == "__main__":
    create_test_data()
    start = time.time() * 1000  # Convert to milliseconds
    bubble_sort(input_data)
    end = time.time() * 1000  # Convert to milliseconds
    print(f"Start time is {start}")
    print(f"End time is {end}")
    print(f"Completed time {end-start} milliseconds")
```
