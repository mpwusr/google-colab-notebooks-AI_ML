```python
import datetime

def total_discounted_amount(*prices, discount=0.0):
  total = 0
  for price in prices:
    total += price
  total_discounted_price = total - total * discount
  return total_discounted_price

today = datetime.datetime.now()
todays_discount=0.35
total_disc_amt=total_discounted_amount(500.00, 900.00, 750.00, 850.00, discount=todays_discount)
print("Todays", today.strftime("%a %d %b"), "current discount is:", "{:.0%}".format(todays_discount))
print("Todays total discounted amount is:",'${:,.2f}'.format(total_disc_amt))
```

```python
def customer_net_spend(*prices, **kwargs):
  total = 0
  for price in prices:
    total += price
  total_discounted_price = total - kwargs['discount']*total
  net_spend = total_discounted_price - kwargs['cashback']
  if total >= 10000:
    reward_points = 300
  elif total >= 5000:
    reward_points = 200
  elif total >= 2000:
    reward_points = 100
  else:
    reward_points = 0
  return total, net_spend, reward_points
today = datetime.datetime.now()
todays_cashback=5
todays_discount=0.35
todays_reward_points = 0
total_disc_amt=total_discounted_amount(1500.00, 1900.00, 2750.00, 850.00, discount=todays_discount)
print("Todays date", today.strftime("%a %d %b"))
print("Todays current discount is:", "{:.0%}".format(todays_discount))
print("Todays total discounted amount is:",'${:,.2f}'.format(total_disc_amt))
additionals = {"cashback": todays_cashback, "discount": todays_discount}
total_amount, today_customer_spend, todays_reward_points=customer_net_spend(1500.00, 1900.00, 2750.00, 850.00, **additionals)
print("Todays customer spend amount is:",'${:,.2f}'.format(today_customer_spend))
print("Todays rewards is:", todays_reward_points)
```
