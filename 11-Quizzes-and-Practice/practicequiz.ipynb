```python
# Libraries to help with reading and manipulating data
import numpy as np
import pandas as pd

# Libraries to help with data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Command to tell Python to actually display the graphs
%matplotlib inline

pd.set_option('display.float_format', lambda x: '%.2f' % x) # To supress numerical display in scientific notations
```

```python
df = pd.read_csv('/content/drive/My Drive/Placement_Data.csv')
```

```python

sns.countplot(x = 'degree_p', data = df, hue = 'status');
```

```python
sns.countplot(x = 'hsc_s', data = df);
```

```python
sns.heatmap(data=df[['ssc_p','hsc_p','degree_p','etest_p', 'mba_p']].corr(), annot=True, cmap='YlGnBu');
```

```python
plt.title('Boxplot:Degree precentage')
plt.xlim(40,100)
plt.xlabel('Degree percentage')
sns.axes_style('whitegrid')
sns.boxplot(data=df, x='degree_p', hue='status');
```

```python
plt.title('Boxplot:Degree precentage')
plt.xlim(50 ,100)
plt.ylim(0,10)
plt.xlabel('Degree percentage')
sns.axes_style('whitegrid')
sns.countplot(data=df, x='ssc_p', hue='status', dodge=False);
plt.show() # this will ensure that the plot is displayed without the text
```
