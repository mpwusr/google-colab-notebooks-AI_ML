```python
power = 1
while power < 5:
  power += 1
  print("@", end="")
  if power == 3:
    break
else:
  print("@")
```

```python
others = -1
for i in range(1, 3):
  for j in range(1, 2):
    if i == j:
      others += 1
  else:
    others += 1
print(others)
```
