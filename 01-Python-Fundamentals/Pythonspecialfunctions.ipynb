Python lambda

```python
str1 = 'Geeks/for/Geeks'
lambdasplitslash = lambda string: string.split('/')
lambdaupper = lambda string: string.upper()
print(lambdasplitslash(str1))
print(lambdaupper(str1))
```

formatting

```python
format_numeric = lambda num: f"{num:e}" if isinstance(num, int) else f"{num:,.2f}"

print("Int formatting:", format_numeric(1000000))
print("float formatting:", format_numeric(999999.789541235))
```

Lambda cube

```python
def cube(y):
    return y*y*y

lambda_cube = lambda y: y*y*y
print("Using function defined with `def` keyword, cube:", cube(5))
print("Using lambda function, cube:", lambda_cube(5))
```

Lambda list

```python
is_even_list = [lambda arg=x: arg * 10 for x in range(1, 5)]
for item in is_even_list:
    print(item())
```

Lambda if.. else

```python
lambdaMax = lambda a, b : a if(a > b) else b
print(lambdaMax(1, 2))
print(lambdaMax(5, 7))
print(lambdaMax(50, 4))
print(lambdaMax(13, 56))
```

sorted List

```python
List = [[2,3,4],[1, 4, 16, 64],[3, 6, 9, 12]]

sortList = lambda x: (sorted(i) for i in x)
secondLargest = lambda x, f : [y[len(y)-2] for y in f(x)]
res = secondLargest(List, sortList)

print(res)
```

filter odd

```python
from typing_extensions import Deque

li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]

odd_list = list(filter(lambda x: (x % 2 != 0), li))
even_list = list(filter(lambda x: (x % 2 == 0), li))
print(f'original list: {li}')
print(f'all odd numbers in list: {odd_list}')
print(f'all even numbers in list: {even_list}')
print(tuple(li))
print(set(li))
print(Deque(li))
```

```python
li = [5, 7, 22, 97, 54, 62, 77, 23, 73, 61]

final_list = list(map(lambda x: x*2, li))
print(final_list)
```
